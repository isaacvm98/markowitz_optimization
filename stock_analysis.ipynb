```python
for i in range(len(portfolios.keys())-1):
    name = list(portfolios.keys())[i]
    print(f"\nAnalyzing portfolio: {name}")
    portfolio, strategy = analyze(f"{name} Stocks", portfolios[name])

    portfolio.plot_weights(strategy['Max Sharpe'], f"{name} - Max Sharpe Portfolio")
    portfolio.plot_weights(strategy['Market Cap'], f"{name} - Market Cap Weights")
    
    # Plot efficient frontier
    print("\nShowing efficient frontier...")
    portfolio.plot_frontier(fname=f"{name} Efficient Frontier")
```
